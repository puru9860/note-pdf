````
import json
import cv2 
import numpy as np

def convert_image_to_text(image_path):

  img = cv2.imread(image_path,0)
  img = np.array(img)
  if img.shape != (1024,1024):
    img = cv2.resize(img,(1024,1024),cv2.INTER_CUBIC)
  cv2.imwrite('detect/image.jpg',img)

  !./darknet/darknet detector test YOLOv3/annotation/obj.data YOLOv3/cfg/yolov3.cfg YOLOv3/backupv2/yolov3_last.weights  detect/image.jpg -out detect/result.json -ext_output -thresh 0.7 -dont_show
  !python3 detect/preprocessing.py
  !./darknet/darknet detector test YOLOv3/cropped/annotation/obj.data YOLOv3/cropped/yolov3.cfg YOLOv3/cropped/backup/yolov3_last.weights  -thresh 0.3 -dont_show   < detect/test.txt> detect/words.txt

def print_converted_text():
  from IPython.display import clear_output 
  result = open('result.txt')
  output = result.readlines()
  words = []
  word = []
  n=0
  word_list = open("detect/output.txt", "w")  
  for line in output:
      if line.endswith('%\n'):
        if line[0] == 'S' and word!= []:
          word.append('s')
        else:
          word.append(line[0])

      else:
          if word != []:
              words.append(''.join(word))
              n+=1

              word = []
  # clear_output()
  words = ' '.join(words)
  word_list.write(words)
  print("total words detected: " + str(n))
  print(words)
  word_list.close()
````

````
import os
import numpy as np
import shutil


src = 'YOLOv3/cropped/data'
word='word/data/train'


allFileNames = os.listdir(src)
fileNames = []
for file in allFileNames:
    if file.endswith('.jpg'):
      if os.path.exists(src+'/'+file.split('.')[0]+'.txt'):
        fileNames.append(file)
np.random.shuffle(fileNames)
train_FileNames, val_FileNames = np.split(np.array(fileNames),[int(len(fileNames)*0.8)])


train_FileNames = [src+'/'+ name for name in train_FileNames.tolist()]
val_FileNames = [src+'/' + name for name in val_FileNames.tolist()]
print('Total images: ', len(fileNames))
print('Training: ', len(train_FileNames))
print('Validation: ', len(val_FileNames))
````
